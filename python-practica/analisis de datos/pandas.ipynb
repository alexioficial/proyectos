```python
# filtrar texto que contenga ...
df[df['descripcion'].str.contains('4')][['descripcion', 'costo']]
```

```python
# Sumando 100 al precio
Sumar100 = lambda x: x + 100

df['precio_mas_100'] = df['precio1'].apply(Sumar100)
df[(df['costo'] != 0) & (df['precio1'] != 0.0)][['descripcion', 'costo', 'precio1', 'precio_mas_100']]
```

```python
# Test
Promediar = lambda nums: sum(nums) / len(nums)
df['promedio'] = Promediar([df[f'precio{i}'] for i in range(1, 4)])
df[(df['costo'] != 0) & (df['precio1'] != 0.0)][['descripcion', 'costo', 'promedio']]
```

```python
import matplotlib.pyplot as plt
df['ganancia'] = df['promedio'] - df['costo']
df['precio'] = df['promedio']
df[(df['costo'] != 0) & (df['precio1'] != 0.0)][['descripcion', 'costo', 'precio', 'ganancia']].plot(kind = 'bar')
plt.show()
```
